```python
df_full = pd.read_csv('Opentable_dataset.csv')
```

```python

list=['District of Columbia', 'Philadelphia', 'Cincinnati', 'Honolulu', 'New York', 'Las Vegas', 'Phoenix', 'Nashville', 'San Francisco', 'Houston', 'Boston', 'Portland', 'Los Angeles', 'Dallas', 'New Orleans', 'Denver', 'Chicago', 'Austin', 'Seattle', 'Atlanta', 'Miami']

df = df_full[df_full.Name.isin(list)]
df
```

```python
#since I am only comparing cities, I am dropping index "45" which is NY state
df.drop([45], inplace=True)
```

```python
df.set_index('Name', inplace = True)
```

```python
df = df.transpose()
```

```python
df = df.drop(["Type"])
```

```python
df = df.set_index(pd.to_datetime(df.index))
```

```python
df.head(30)
```

```python
df = df[df.columns].replace('[%]', '', regex=True).astype(float)/100
```

```python
df.head(30)
```

```python
df_2 = df[['New York', 'Los Angeles', 'Dallas', 'New Orleans', 'Miami', 'Phoenix']].copy()
df_2
```

```python
df_new = df_2.resample('H').median()
```

```python
df_new.head(10)
```

```python
df_new = df_new.interpolate(method='linear')
```

```python
df_new.head()
```

```python
adf_results = {}
for col in df_new.columns.values:  
    adf_results[col] = tsa.adfuller(df_new[col])

adf_results
```

```python
# loop over dictionary data
non_stationary = []
stationary = []
for key, value in adf_results.items():
    if value[1] > 0.05:
        non_stationary.append(key)
    else:
        stationary.append(key)
print (non_stationary)
print (stationary)
```

```python
cutoff = round(NY.shape[0]*0.8)
cutoff
```

```python
NY_train = NY[:cutoff]
NY_test = NY[cutoff:]
```

```python
#look at the data
NY_train.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(NY_train, label='train')
ax.plot(NY_test, label='test')
ax.set_title('NY Train-Test Split')
plt.legend();
```

```python
# Build Model
model = ARIMA(NY_train, order=(2,0,6))  
new_model = model.fit()  
print(new_model.summary())
```

```python
y_hat_train = new_model.predict(typ='levels')
y_hat_test = new_model.predict(start=NY_test.index[0], end=NY_test.index[-1],typ='levels')

fig, ax = plt.subplots()
ax.plot(NY_train, label='train')
ax.plot(NY_test, label='test')
ax.plot(y_hat_train, label='train_pred')
ax.plot(y_hat_test, label='test_pred')

plt.legend();
```

```python
# Let's zoom in on test
fig, ax = plt.subplots()

ax.plot(NY_test, label='true')
ax.plot(y_hat_test, label='pred')

plt.legend();
```

```python
np.sqrt(mean_squared_error(NY_test, y_hat_test))
```
